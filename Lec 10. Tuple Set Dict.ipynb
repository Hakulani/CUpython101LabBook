x.sort() x ต้องเป็น list เท่านั้น แต่ sorted(x) เป็นอะไรก็ได้

```python
def sorted(x): #นำข้อมูลใน X มาเรียงลำดับข้อมูล แล้วคืนมาเป็น list ที่เรียง สว่น xยังเหมือนเดิม
    out=[]
    for e in x:
        out.append(e)
    out.sort()
    return out
x=sorted([1,4,3,2])
print(x)
```

ตรวจสอบข้อมูลซ้ำกันใน list

```python
#วิธีช้าไล่เช็คทุกคู่
def has_duplicate(x):
    for i in range(len(x)-1):
        for j in range(i+1,len(x)):
            if x[i]==x[j]:
                return True
    return False
has_duplicate([2,3,3,4,5,6,6,6,7,8,8,8,9])
            
```

```python
#วิธีเร็วขึ้น  เอาxมา sort ก่อน ตัวที่เหมือนกันจะอยู่ใกล้กัน เทียบทีละคู่ ถ้าเหมือนก็จบ
def has_duplicater(x):
    d=sorted(x) #สร้าง list ใหม่
    print(d)
    for i in range(len(x)-1):
        if d[i]==d[i+1]:#เทียบทีละคู่
            return True
    return False
has_duplicater([2,3,3,4,5,6,6,6,7,8,8,8,9])            
```

ตรวจสอบข้อมูลซ้ำกันใน list โดยใช้ set

```python
#จำนวนใน list x กับ ใน set ไม่เท่ากัน
def has_duplicateset(x):
    s=set(x)
    return len(s)!=len(x)
has_duplicateset([2,3,3,4,5,6,6,6,7,8,8,8,9])   
```

```python
#สร้าง set ว่างก่อน อ่าน x ทีจะตัวแล้วเพิ่มไปใน set s แต่ก่อนจะเพิ่มตามว่า eอยู่ใน set x หรือเปล่าถ้าอยู่ ก็ return True
def has_duplicatesets(x):
    s=set(x)
    for e in x:
        if e in s:
            return (True,s)
        s.add(e)
 
    return False
has_duplicatesets([2,3,3,4,5,6,6,6,7,8,8,8,9])   
```

ตัวอย่าง หาสองจำนวนต่างกกันที่รวมกันได้ k

```python
# 1 2 3 4 8 9 -8 -2
x=[int(e) for e in input().split()]
d=set(x) #สร้าง set ไม่มีตัวซ้ำ
k=int(input())
soln=[]
for e in d:
    e1 = k- e # k=11  หาว่าอะไรก็ได้บวก e แล้วได้ 11 , 
    if e<e1 and e1 in d:  #e1 อยู่ใน set d ไหม และ ป้องกันตัวซ้ำกัน เช่น (2,9)กับ (9,2)  e<e1 
        soln.append((e,e1)) #เก็บเป็น tuple [(2,9),(3,8)]
if len(soln)==0: #ตรวจสอบว่ามีสมาชิกรึเปล่า
    print('Not Found')
else:
    print (soln)
    
```

ต้องการสร้าง dict จาก input

```python
#รูปแบบ {str:set}  {key:value}
cartoon={}
x = input()
while x!='q':
    name,atype = x.split(",") #รับ input 2 ตัว  ชื่อ , ประเภท
    if atype not in cartoon: #ถามก่อนว่าไม่อยู่ใน cartoon
        cartoon[atype]={name} #ยังไม่มีสร้างset ใหม่ เพิ่มไปใน dict  ห้ามใช้ set(name) จะกลายเป็น n,a,m,e
    else:
        cartoon[atype].add(name) #ถ้ามีอยู่แลว้เพิ่มไปใน set เดิม  เพิ่มไปใน value
    x=input()
print(cartoon)
```

การแจกแจง Dict

```python
D={"Vios":"Toyota" , "Fortuner":"Toyota",
  "Wave":"Honda","Civic":"Honda"}
#แจกแจง key
print("for k in D.keys():")
for k in D.keys():
    print(k)
#แจกแจง values
print("\nfor k in D.values():")
for k in D.values():
    print(k)
#แจกแจง key,values    
print("\nfor k inD.tiems():")
for k,v in D.items():
    print(k,v)
```

revers mapping

```python
#กรณี value ไม่ซ้ำกัน
def reverse(d):
    r={}
    for k,v in d.items():
        r[v]=k
    return r
```

```python
#กรณี value  ซ้ำกัน
d={"Vios":"Toyota" , "Fortuner":"Toyota",
  "Wave":"Honda","Civic":"Honda"}
def reverses(d):
    r={}
    for k,v in d.items():
        if v not in r: #ถ้า v ไม่อยู่ใน r สร้าง set ว่าง
            r[v]=set() # v เป็น key
        r[v].add(k) # k เป็น value
    return r
reverses(d)
```
