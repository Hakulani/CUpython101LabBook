ข้อมูลนำเข้า สตริงหนึ่งบรรทัด

````python
lst=list('0123456789')
s=input() #list of string
for e in s:
    if e in lst:
        lst.remove(e) 
if len(lst)==0:
    print("None")
else:
    print(','.join(lst))
    
````

05-04: การนับจำนวนยอด  
ถ้านำข้อมูลในลิสต์ y = [1, 4, 5.1, 5.2, 5.3, 6, 9, 6.1, 6.2, 7.1, 7.2, 7.3, 8.4, 8.5, 9.2, 9.3, 20, 6, 8, 11.2, 11.5] ไปวาดกราฟ
เส้นจะได้ดังรูปข้างล่างนี้ จะเห็นว่ามีการเปลี่ยนแปลงของข้อมูลที่มองได้ว่าเป็น "ยอด" คือ ข้อมูลตัวที่เป็นยอด มีค่ามากกว่าทั้งตัว
ติดกันทางด้านซ้ายและด้านขวา จงเขียนโปรแกรมที่รับรายการของจำนวน แล้วแสดงว่าลิสต์นี้มีอยู่กี่ยอด

Input 1 1 1 1 1 1 9 1 1 1 1 1</br>
Output 1
Input 1 9 1 9 1 9 1 9 1 9 1</br>
Output 5

````python
y=input()
count=0
y=[float(x) for x in y.split()]
for e in range (1,len(y)-1) :
    
    if (y[e-1]<y[e]) and (y[e]>y[e+1]):
        count+=1
print(count)
````

ข้อแนะนำ: หลังจากอ่านข้อมูลเก็บใส่ลิสต์แล้ว ให้ใช้ sort เรียงลำดับข้อมูลในลิสต์จากน้อยไปมาก จะทำให้หาจำนวนข้อมูลที่ต่างกัน</br>
ง่ายขึ้น ด้วยการดูข้อมูลทีละตัวจากซ้ายไปขวาเพียงรอบเดียวเท่านั้น เช่น จากข้อมูลตัวอย่าง 11, 12, 13, 11, 13, 13, 13, 13, 11</br>
หลังจากเรียงลำดับแล้วจะได้ 11, 11, 11, 12, 13, 13, 13, 13, 13 จะเห็นได้ว่า จำนวนค่าที่ต่างกันก็จะเท่ากับจำนวนคู่ที่ติดกันที่มี</br>
ค่าไม่เท่ากัน บวกอีกหนึ่ง</br>
ข้อมูลนำเข้า</br>
รายการของจำนวนเต็ม</br>
ข้อมูลส่งออก</br>
บรรทัดแรกบอกว่ามีค่าที่ต่างกันกี่ตัว</br>
บรรทัดที่สอง แสดงลิสต์ของค่าที่ต่างกัน (เรียงจากน้อยไปมาก) ขอแค่ 10 ตัวแรกก็พอ (ถ้ามีไม่ถึงก็เอาเท่าที่มี)</br>

````python
x=input()
d=[int(i) for i in x.split()]
unique=[d[0]]
for  i in range(1,len(d)):
    if d[i]!=d[i-1]:
        unique.append(d[i])
print(len(unique))
print(unique[:10])
````

ปัญหา Collatz ถามว่า ถ้าเริ่มที่จำนวนเต็มบวก n การเปลี่ยนแปลงค่า n ด้วยวิธีข้างล่างนี้จะทำให้ในที่สุดแล้ว n เป็น 1 หรือไม่
```
while n != 1:
    if n is even: #ถ้า n เป็นเลขคู่ n%2==0
        n = n // 2 # ปัดเศษทิ้ง
    else:
        n = 3*n + 1
```

````python

n=int(input())
lst=[str(n)]
while n!= 1:
    if  n%2==0:
        n = n //2 # ปัดเศษทิ้ง 
    else:
        n = 3*n+1
    lst.append(str(n)) 

print('->'.join(lst[-15:])) #เริ่มที่ -15
        
````

05-07: การปรับเกรด  
จงเขียนโปรแกรมอ่านรายการของเลขประจำตัวนักเรียนและเกรด ตามด้วยรายการของเลขประจำตัวที่ได้รับการปรับเกรดอีกหนึ่ง
ประจุ (จาก B+ เป็น A, จาก B เป็น B+, จาก C+ เป็น B, ... )

### ข้อมูลนำเข้า 
แบ่งเป็นสองส่วน</br>
- ส่วนแรกมีหลายบรรทัดเป็นรายการของเลขประจำตัวนักเรียนและเกรด บรรทัดละหนึ่งคน ปิดท้ายบรรทัดสุดท้ายด้วย
ตัวอักษร q
- ส่วนที่สองมีบรรทัดเดียวเป็นรายการของเลขประจำตัว (คั่นด้วยช่องว่าง) ที่ได้รับการปรับเกรดหนึ่งประจุ</br>

### ข้อมูลส่งออก</br>
รายการของเลขประจำตัวนักเรียนและเกรด (หลังได้รับการปรับเกรดแล้ว) ของทุกคน คนละบรรทัด ในลำดับเดียวกับที่อ่านในข้อมูล</br>
นำเข้า

````python
ids = []
grades = []
uids = []
level=['F','D','D+','C','C+','B','B+','A']
x=input()
while  x !='q':
    sid,g=[e for e in x.split()]
    ids.append(sid)
    grades.append(g)
    x=input()
uids=input().split()
for e in uids:
    if e in ids:
        i=ids.index(e)
        if grades[i] in level :
            k=level.index(grades[i])
            if grades[i] !='A':
                grades[i]=level[k+1]        
for i in range(len(ids)):
    print(ids[i],grades[i])
````

05-10: ตัดและกรีด 
จงเขียนโปรแกรมรับลำดับของไพ่ในสำรับ จากนั้นนำไพ่สำรับนี้มา "ตัด" กับ "กรีด" ตามที่กำหนด
นิยามการตัดไพ่ คือการแบ่งไพ่ในสำรับออกเป็นสองกอง กองละเท่ากัน จากนั้นนำกองหลังมาทับกองหน้า เช่น

ข้อมูลนำเข้า</br>
บรรทัดแรกเป็นลำดับของชื่อไพ่จากบนลงล่างในสำรับ (ไม่จำเป็นต้องมี 52 ใบ) เป็นสตริงคั่นด้วยช่องว่าง (ประกอบด้วยสตริงเป็น</br>
จำนวนคู่)</br>
บรรทัดที่สองเป็นลำดับการตัดหรือกรีดไพ่ เป็นสตริงที่ประกอบด้วยตัวอักษร ถ้าเป็น C คือให้ตัดไพ่ ถ้าเป็น S คือให้กรีดไพ่</br>
ถ้าเป็นอักษรอื่นไม่ต้องทำอะไร</br>

ข้อมูลส่งออก</br>
ชื่อไพ่ทั้งหมดจากบนลงล่างในสำรับ หลังจากตัดหรือกรีดตามที่กำหนด</br>

````python
s=input().split()
cmd=input()
n= len(s) 
for c in cmd:
    if c=='C': #cut
        s=s[n//2:]+s[:n//2] #สลับหน้าหลัง
    elif c =='S': #shuffle
        
        lst=[' ']*n # Empy list of string
        lst[::2]= s[:n//2]#ครึ่งแรกวางตำแหน่งคู่  เริ่มที่ 0
        lst[1::2]= s[n//2:] #ครึ่งหลังวางตำแหน่งคี่ เริ่มที่ 1
        s=lst
print(' '.join(s))
````

````python
queue=[]
itcketID=0
sum_qtime=0
nocustomer=0
n = int(input()) # อ่านจานวนคาสั่ง
for k in range(n):
    cmd = input().split() # อ่านข้อมูลคำสั่ง
    if cmd[0] == 'reset':
        ticketID=int(cmd[1])
    elif cmd[0] == 'new':
        t =int(cmd[1])
        queue.append([ticketID,t])
        print('ticket',ticketID)
        ticketID+=1
    elif cmd[0] == 'next': #เรียนลูกค้าหัวคิวออกมา
        next_id,queue_t=queue.pop(0)
        print('call',next_id)
    elif cmd[0] == 'order':#จดออเดอร์  เวลารอคอย ordert- queuet(next<new)
        order_t = int(cmd[1])
        dt=order_t-queue_t
        print('qtime',next_id,dt)
        nocustomer+=1
        sum_qtime+=dt
 
    elif cmd[0] == 'avg_qtime': #เวลารอเฉลี่ยของลูกค้าที่ได้รับ order
        print('avg_qtime' , round((sum_qtime/nocustomer),4) )
````
