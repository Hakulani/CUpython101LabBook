‚Ä¢ x = ‡∏à‡∏∏‡∏î‡∏Å‡∏∂‡πà‡∏á‡∏Å‡∏•‡∏≤‡∏á‡∏Ç‡∏≠‡∏á‡∏ä‡πà‡∏ß‡∏á

```python
import math
a=float(input())
L=0
U=a
x=(U+L)/2
#while not math.isclose(a,x**2):
while (abs(a-x**2)>1e-10*max(a,x**2)):
    print('x=',x)
    if x**2>a:
        U=x
    else:
        L=x
    x=(U+L)/2

print(x)
```

```python
n=int(input())
c=0
red=[0]*n #‡∏™‡∏£‡πâ‡∏≤‡∏á list ‡∏£‡∏≠ n ‡∏ä‡πà‡∏≠‡∏á
blue=[0]*n
for i in range(n):
    x,y=(input().split())
    x=int(x)
    y=int(y)
    if i%2==0:
        red[i]=x
        blue[i]=y
    else:
        red[i]=y
        blue[i]=x
zz=input()
if zz =='Zig-Zag':
    print(min(red),max(blue))
else:
    print(min(blue),max(red))
```

```python
x,y=(input().split())
x,y=int(x),int(y)
min_zigzag = x
max_zigzag = y
min_zagzig = y
max_zagzig = x
i=1 #‡πÄ‡∏£‡∏¥‡πà‡∏°‡πÅ‡∏ñ‡∏ß‡∏ó‡∏µ‡πà 1 ‡πÄ‡∏û‡∏£‡∏≤‡∏∞‡∏Ñ‡πà‡∏≤‡πÅ‡∏£‡∏Å‡∏Ñ‡∏∑‡∏≠‡πÅ‡∏ñ‡∏ß‡∏ó‡∏µ‡πà 0 ‡πÅ‡∏ñ‡∏ß‡∏ó‡∏µ‡πà 1 ‡∏à‡∏∞‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏î‡πâ‡∏ß‡∏¢ max x2, min y2
while True:
    d=input().split()
    if len(d)==1:break  #‡∏ñ‡πâ‡∏≤‡∏û‡∏¥‡∏°‡∏û‡πå Zig-Zag ‡∏´‡∏£‡∏∑‡∏≠ Zag-Zig ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏à‡∏∞‡πÄ‡∏´‡∏•‡∏∑‡∏≠‡∏Ñ‡πà‡∏≤‡πÄ‡∏î‡∏µ‡∏¢‡∏ß
    x,y=int(d[0]),int(d[1])
    if i%2==1:#‡∏Ñ‡∏µ‡πà‡∏™‡∏•‡∏±‡∏ö
        x,y=y,x
    min_zigzag =min(min_zigzag,x)
    max_zigzag =max(max_zigzag,y)

    min_zagzig =min(min_zagzig,y)
    max_zagzig =max(max_zagzig,x)

    i+=1

zz=d[0]
if zz=='Zig-Zag':
    print(min_zigzag,max_zigzag)
else:
    print(min_zagzig,max_zagzig)
```
