```python
#check min max
min_x=int(input())
k=0
while k<4:
    x=int(input())
    if x< min_x:
        min_x=x
    k+=1
print('min=',min_x)
```

```python
#check min max
n=int(input()) #กำหนดว่าจะรับกี่ตัว
min_x=float(input())
k=0
while k<n-1:
    x=float(input())
    if x< min_x:
        min_x=x
    k+=1
print('min=',min_x)
```

```python
#check min max
#รับค่าจนกว่าจะมีตัวพิเศษ q
min_x=int(input()) #รับตัวแรก
e=input() #รับตัวที่ 2 string
while e!="q":
        x=float(e)
        if x< min_x:
            min_x=x
        e=input()
print('min=',min_x)
```

```python
#check เลข 0 และนับว่ามีกี่ตัว
x=input().split()
i=0
count=0
while i<len(x):
    if int(x[i])==0:
        count+=1
    i+=1
print(count)
```

หารากที่ 2 ด้วยBisection 

```python
a = float(input())
L=0
U=a
count=0
x=(L+U)/2
while abs(x**2-a)>1e-10:
    if x**2>a:
        U=x
    else:
        L=x
    x=(L+U)/2
    count+=1
print(x, "round=",count)
```

```python
a = float(input())
L=0
U=a
count=0
x=(L+U)/2
#while abs(x**2-a)/max(x**2,a)>1e-10:
while abs(x**2-a)>1e-10*max(x**2,a):
    if x**2>a:
        U=x
    else:
        L=x
    x=(L+U)/2
    count+=1
print(x, "round=",count)
```

```python
#remove special
s=input()
t=""
for ch in s:
    if"0"<=ch<="9"or \
    "A"<=ch<="Z" or \
    "a"<=ch<="z" or\
    ch==" ":
        t+=ch #เอาตัวที่สนใจไปเติมใน t
print(t)
```
