02-04: นำหน้าด้วย 0<br>
จงเขียนโปรแกรมรับจำนวนเต็มบวก M กับ N เพื่อแสดงค่า M ทางจอภาพ โดยถ้าค่า M มีจำนวนหลักน้อยกว่า N ก็ให้เติม 0 ด้านซ้าย
ให้จำนวนหลักทั้งหมดครบ N ตัว เช่น ถ้า M=123, N=5 จะแสดง 00123 ถ้า M=12345, N=3 จะแสดง 12345<br>
ข้อมูลนำเข้า จำนวนเต็มบวก 2 จำนวน สำหรับ M กับ N จำนวนละบรรทัด
- `123`
- `5`<br>
ข้อมูลส่งออกค่าของ M ที่อาจมีการเติม 0 ด้านซ้ายเพื่อให้มีจำนวนหลักอย่างน้อย N หลัก
- `00123`<br>
ข้อแนะนำ: Python มี built-in function ชื่อ max มีไว้ใช้หาค่ามากสุดของข้อมูล เช่น max(5,9) ได้ เพราะมากสุดคือ 9

```python
M=input('M:')
N=int(input('N:'))

if len(M)>N:
    print(M)
else :
    print(str(0)*(N-len(M))+M)
```

02-08: จำนวนทศนิยมและเศษส่วน<br>
เราสามารถเขียนจำนวนตรรกยะในรูปแบบเศษส่วนหรือแบบทศนิยมได้ <br>
เช่น 1/8 = 0.125 แต่ก็มีจำนวนตรรกยะที่เขียนออกมาได้เป็น<br>
เลขหลังจุดทศนิยมไม่รู้จบแบบซ้ำ เช่น 3221/555 = 5.8036036036036036... (เลข 036 จะซ้ำไปเรื่อย ๆ ไม่รู้จบ) ในกรณีนี้ ขอเขียน<br>
เป็น 5.8(036) แสดงให้เห็นว่า เลขในวงเล็บ 036 จะซ้ำไม่รู้จบ จงเขียนโปรแกรมที่รับจำนวนในรูปแบบทศนิยม แล้วแสดงในรูปแบบ<br>
เศษส่วน<br>
ข้อมูลนำเข้า<br>
จำนวนไม่ติดลบแบบทศนิยม ที่แบ่งทศนิยมเป็นสามส่วนคั่นด้วยจุลภาคคือ เลขหน้าจุด เลขหลังจุดที่ไม่อยู่ในวงเล็บ และเลขในวงเล็บ
(ดูตัวอย่าง)<br>
ข้อมูลส่งออก
จำนวนในรูปแบบเศษส่วนที่มีค่าเดียวกับจำนวนที่รับเข้ามา โดยที่ค่าของเศษและส่วนมี ห.ร.ม. เป็น 1 (ดูตัวอย่าง)<br>

`10.234567856785678... 10,234,5678`  a,b,c
        x = 10.234567856785678...  
    10**3x = 10234.5678
10**7x = 102345678.5678
10**7x-10**3x = 102345678-10234
x= (102345678-10234)/(10**7-10**3)
nb=len(b) , nc=len(c)
x = int(a+b+c) (ต่อ string) - int(a+b) / (10**(nb+nc)-10**nb)

```python
import math
a,b,c = input().split(',')
nb=len(b)
nc=len(c)
n=int(a+b+c)-int(a+b)
d=10**(nb+nc)-10**nb
g = math.gcd(n,d)# หรม 
print('gcd.=',g)
print(n ,'/',d )
print(n//g,'/',d//g)
```
