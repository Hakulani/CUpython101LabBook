หยิบข้อมูลใน list มาประมวลผล

```python
#KTB 18.9 BBL 176.5 SCB 132.0 KBANK 172.0 TMB 1.8 อยากสร้าง 2 list เก็บชื่อย่อ กับ ราคา
x=input().split()
#list เก็บชื่อย่อ 
s=x[::2]#หยิบเฉพาะตำแหน่งคู่ (เริ่มที่ 0) 0,2,4,...
#list เก็บราคา
p=[]
for e in x[1::2]: #หยิบเฉพาะตำแหน่งคี่ (เริ่มที่ 1) 1 , 3 ,5 ,...
     p.append(float(e))
p0=float(input()) #ใส่ราคาที่อยากค้นหา

for i in range (len(p)): # วนหาข้อมูล ต้องการรู้ข้อมูลและตำแหน่ง
    if p[i] < p0: #หาตัวที่ราคาต่ำกว่า P0 ใน p
        print(s[i])
        

 
```

```python
#ตัวที่ติดกันเท่ากัน
d=[1,2,3,4,5,5,3]
found = False
for i in range(1,len(d)):
       if d[i-1]==d[i]:
            found=True
            print(d[i-1],d[i],'index',i-1,i)
            break
print(found)
```

แบบ List of list  [['KTB', 18.9], ['BBL', 176.5], ['SCB', 132.0], ['KBANK', 172.0], ['TMB', 1.8]]

```python
#KTB 18.9 BBL 176.5 SCB 132.0 KBANK 172.0 TMB 1.8     list เก็บ[ชื่อย่อ กับ ,ราคา]
x=input().split()
s=[]
q=input()
for i in range (0,len(x),2):  #ขยับที่ละ 2
    s.append([x[i],float(x[i+1])])

found=False
for e in s:
    if q ==e[0]: #[0,1]   ['KTB', 18.9]
        print(e[1])
        found=True
        break
#if found==False:
if not found:
    print("Not found")
        
```

```python
#KTB 18.9 BBL 176.5 SCB 132.0 KBANK 172.0 TMB 1.8     list เก็บ[ชื่อย่อ กับ ,ราคา]
x=input().split()
s=[]
q=input()
for i in range (0,len(x),2):  #ขยับที่ละ 2
    s.append([x[i],float(x[i+1])])

found=False
for [name,price] in s: #list ที่มี 2ช่อง [name,price]
    if q ==name:
        print(price)
        found=True
        break
#if found==False:
if not found:
    print("Not found")
```
