```python
def days_in_feb(y):
    if y%400==0 or ( y%4==0 and y%100!=0):
        return 29
    else:
        return 28
def days_in_month (m,y):
    if m==2:
        return days_in_feb(y)
    if m==[4,6,9,11]:
        return 30
    else:
        return 31
def days_in_year(y):
    return 365-28+days_in_feb(y)
```

```python
def count (x,a,b): #รับ list x มีข้อมูลที่อยู่ในช่วง [a,b] มีกี่ตัว
    c=0
    for e in x:
        if a<=e<=b:
            c+=1
    return c
d=[1,2,3,4,4,34,3,2,4,22]
print(count(d,2,6))
```

```python
def count_match(x,a): #รับ list x มีข้อมูลที่ =a มีกี่ตัว
    c=0
    for e in x:
        if e==a:
            c+=1
    return c
d=[1,2,3,4,4,34,3,2,4,22]
print(count_match(d,4))
```

```python
def mode(x):
    max_c=count_match(x,x[0]) #ตัวตัวแรกว่าปรากฏกี่ครั้ง
    max_e=x[0] 
    for e in x[1:]: #ตัวที่ 1เป็นต้นไป
        c=count_match(x,e)
        if c > max_c: #ถ้าเจอตัวมากกว่า
            max_c=c #update c
            max_e=e #เปลี่ยนตัวนั้น เป็น e max_e
    return max_e
d=[1,2,3,4,4,34,3,2,4,22]
print(count_match(d,2))
print('mode=',mode(d))         
        
```

```python
#ความยาวเส้นรอบรูป
def distance(p1,p2):
    dx=p1[0]-p2[0]
    dy=p1[1]-p2[1]
    return (dx**2+dy**2)**0.5

def perimeter(points):
    p=0
    for i in range(len(points)):
        p+= distance(points[i],points[(i+1)%len(points)]) #ป้องกันไม่ให้เกิน 5ตัว  เช่น 5 mod 5 = 0
    return p
points = [[226,77],[283,121],[337,64],[374,147],[288,218],[198,164]]
pr =perimeter(points)   
print(pr)
```

```python
#รับ list คืน list
import math
def sin(x):
    y=[]
    for e in x:
        y.append(math.sin(e))
    return y #คืน list
x1=[]
for i in range(10):
    x1.append(math.pi*(i)/10) #เอา pi หาร 10จุด  0ถึง 180องศา
y1=sin(x1) #รับ list
print(y1)
```

```python
def roots (a,b,c):
    t=(b**2-4*a*c)**0.5
    r1=(-b+t)/(2*a)
    r2=(-b-t)/(2*a)
    return [r1,r2]
x1,x2 = roots(6,-6,-36)
print(x1,x2)
```

```python
#กรณีมีการเปลี่ยนแปลงค่า ใน defให้สร้างใหม่ ทับของเดิม
def abs_all(data): #รับ x
    d=[]
    for e in data:
        d.append(abs(e))
    return d #คืน d
x=[1,-2,3,-4,-5]
x=abs_all(x)
print(x)
```
