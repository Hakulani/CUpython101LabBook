บรรทัดที่สอง แสดงขนาดของเซตที่เป็นผลลัพธ์ของการ intersection ทุกเซต</br>

````python
n=int(input())
su=set([int(e) for e in input().split()])
si=set(su)
for k in range(n-1):
    s=set([int(e) for e in input().split()]) #สร้าง set
    su |= s # union
    si &=s  # intersect
print(len(su))
print(len(si))
   
         
        
````

โปรแกรมข้างล่างนี้ใช้ dict ชื่อว่า R มี key เป็นชื่อ และ value เก็บ set ของชื่อที่ key รู้จัก ดังนั้น จากข้อมูลข้างบนนี้ย่อมได้</br>

จงเชียนฟังก์ชัน knows, is_celeb และ find_celeb ข้างล่างนี้ให้สมบูรณ์ตาม comment และเก็บข้อมูลที่อธิบายข้างบนนี้</br>

## celeb บุคคลที่ทุกคนในงานรู้จักเขา แต่เขาไม่รู้จักคนอื่นในงานเลย

````python
def knows(R,x,y):#x รู้จัก y ไหม หรือ y อยู่ใน R[x]ไหม
    # return True if x knows y
    return y in R[x]
def is_celeb(R,x): # return True if a is celeb, otherwise return False
    #เช็คก่อนว่า X รู้จัก1คน และไม่ใช่ตัวเอง ก็จะไม่ใช่ celeb
    if len(R[x])==1 and R[x] != {x} or len(R[x])>1: #หรือ x รู้จักเกิน 1คน
        return False
    c=sum([1 for e in R if e!=x and x  in R[e]]) #แล้วทุกคนใน R รู้จัก x  ถ้า x in R[e] และไม่ใช่ตัวเอง e!=x เติม 1
    return c==len(R)-1 #คนที่เหลือรู้จักเขาทั้งหมด หักตัวเองออก

def find_celeb(R):
    for e in R:
        if is_celeb(R,e): #ใช่ celebไหม ถ้า True return
            return e
    return None
def read_relations() :
    # build a dictionary R from inputs
    # whose structure is shown in the example
    R = dict()
    while True:
        d = input().split()
        if len(d) == 1 : break
        u1,u2=d
        if u1 not in R: #เช็คก่อนว่า u1 ไม่มีอยู่ใน r
            R[u1]=set() #สร้าง set ว่างเช้าไป
        if u2 not in R: #เช็คก่อนว่า u2 ไม่มีอยู่ใน r
            R[u2]=set() #สร้าง set ว่างเช้าไป   
        R[u1].add(u2) #U1 รู้จัก U2 หรือ u2อยู่ใน set ของคนที่ u1รู้จัก
    return R


def main():
    R = read_relations()
    c = find_celeb(R)
    if c == None :
        print('Not Found')
    else:
        print(c)
exec(input().strip()) # do not remove this line
````

10-08: การบวกและการคูณพหุนาม


โจทย์ข้อนี้เกี่ยวกับการหาผลบวกของ polynomial แบบตัวแปรเดียว 2 ชุด เราสามารถแทน polynomial ด้วยลิสต์ของทูเปิล แต่</br>
ละทูเปิลมีสมาชิก 2 ตัว ตัวแรกคือสัมประสิทธิ์ ตัวหลังคือเลขชี้กำลัง เช่น 4x2 + 3x – 1 แทนด้วย [ (4, 2), (3, 1), (-1, 0) ] โดย</br>
เก็บทูเปิลในลิสต์เรียงลำดับตามเลขชี้กำลังจากมากไปน้อย จงเขียนฟังก์ชัน add_poly(p1, p2)และ mult_poly(p1,</br>
p2) ที่คืนผลบวก และผลคูณของ polynomial p1 กับ p2 ในโครงของโปรแกรมข้างล่างนี้</br>

````python
def add_poly(p1, p2):#บวกพจน์ที่มีเลขยกกำลังเท่ากัน
    p = dict([(e,c) for c,e in p2])# แปลงเป็น dict  {ยกกำลัง:สัมประสิทธิ์}
    for c,e in p1:#หยิบ ทีละตัวใน p1 (c,e)  
        if e in p: # เช็คว่า e อยู่ใน p ไหม (เลขยกกำลังเท่ากันไหม)
            p[e]+=c #บวกพจน์ที่มีเลขยกกำลังเท่ากัน
        else: #กรณีที่หาไม่เจอ ให้เพิ่มตัวใหม่
            p[e]=c #เพิ่ม แทนด้วยdictใน p
#retrun polonimial ในรูป List of tuple เรียงเลขยกกำลังสูงไปต่ำ ถ้าหักล้าง กันได้  coefi  , p[e]=0 ไม่ต้องเอามารวม เช่น 0x
    return [(p[e],e) for e in sorted(p)[::-1] if p[e]!=0] #sort , reverse 
                     
def mult_poly(p1, p2):#คูณ
    p={}
    for c1,e1 in p1:
        for c2,e2 in p2:
             e=e1+e2 #เลขยกกำบัวบวกกัน
             c=c1*c2 #สัมประสิทธิ์คูณกัน
             if e in p: # เช็คว่า e อยู่ใน p ไหม (เลขยกกำลังเท่ากันไหม)
                p[e]+=c #ถ้าใช่ บวกพจน์ที่มีเลขยกกำลังเท่ากัน
             else:#กรณีที่หาไม่เจอ ให้เพิ่มตัวใหม่
                 p[e]=c
    
#retrun polonimial ในรูป List of tuple เรียงเลขยกกำลังสูงไปต่ำ 
    return [(p[e],e) for e in sorted(p)[::-1] if p[e]!=0] #sort , reverse              
             
             
             
for i in range(3):
             exec(input().strip())
````

บรรทัดแรก ระบุจานวนเต็ม n แทนจานวนข้อมูลนิสิต</br>
n บรรทัดถัดมา ระบุข้อมูลนิสิตแต่ละคน ประกอบด้วย ชื่อเล่น กรุ๊ป รุ่น และภาควิชา (รับประกันว่า ชื่อเล่นจะไม่ซ้ำกัน)</br>
บรรทัดสุดท้าย ระบุ กรุ๊ป รุ่น และ/หรือ ภาควิชา ที่จะใช้ค้นหาข้อมูลนิสิต (อาจระบุไม่ครบและไม่เรียงตามลาดับ)</br>

ข้อมูลส่งออก</br>
แสดงข้อมูลนิสิตที่อยู่ในกรุ๊ป รุ่น และภาควิชาที่กำหนด โดยเรียงลาดับชื่อเล่นตามพจนานุกรม ถ้าไม่มีให้แสดงว่า Not Found</br>

````python
#ใช้ list เก็บ แต่เวลาเปรียบเทียบแปลงเป็น set
info=[]
n=int(input())
for k in range(n):
  info.append(input().split()) # เก็บเป็น list ซ้อน list [['Krit', 'A', '97', 'CP'],....]
q=set(input().split()) # รับคำที่ต้องการจะค้น เปรียบเทียบว่าเป็น subset ไหม จึงต้องแปลงเป้น set ก่อน 
r =[]# ถ้ามีการเรียงลำดับ สร้าง list ,เก็บผลลัพธ์
for e in info:  # e เป็น list  ['Krit', 'A', '97', 'CP']     ['Oat', 'A', '98', 'CE']
    if q<=set(e[1:]): #เช็คว่า q เป็น subset ของ e[1:] รึเปล่า
        r.append(e) # เพิ่ม list เข้าไปใน r
#เรียงลำดับ     
if len(r)==0:
    print('Not Found')
else:
    r.sort()
    for e in r: # e list of string  ต้อง ' '.join(e)
        print(' '.join(e))
````

10-10: การเลือกภาควิชา </br>
เมื่อนิสิตเรียนจบชั้นปีที่ 1 นิสิตจะต้องทำกำรเลือกภำควิชำที่จะเรียนต่อในชั้นปีที่ 2</br>
โจทย์ข้อนี้จำลองวิธีที่ใช้ในการเลือกภาควิชาอย่างง่าย (ไม่ตรงกับวิธีที่ใช้จริง) โดยมีขั้นตอนดังนี้</br>
1. แต่ละภาควิชาระบุจำนวนนิสิตที่สามารถรับได้</br>
2. นิสิตแต่ละคน มีรหัสนิสิต คะแนนซึ่งคำนวณจากเกรดของทุกวิชาในชั้นปีที่ 1 และภาควิชาที่เลือก คนละ 4 ภาควิชา เรียง</br>
ตามความต้องการลำดับที่ 1-4 (โจทย์ข้อนี้กำหนดให้ทุกภาควิชาใช้คะแนนเดียวกัน โดยไม่สนใจตัวคูณของแต่ละภาควิชา)</br>
3. นิสิตที่มีคะแนนสูงสุดจะได้เลือกภาควิชาก่อน โดยจะได้เรียนในภาควิชาอันดับสูงที่สุดที่ยังมีที่ว่างอยู่ เช่น หากนิสิตเลือกภาค</br>
PE ME CP MT แต่ภาค PE และ ME รับนิสิตครบจำนวนแล้ว นิสิตจะได้เรียนในภาค CP</br>
4. หลังจากกำหนดภาควิชาให้นิสิตครบแล้ว จะแสดงผลภาควิชาที่นิสิตแต่ละคนได้ เรียงตามรหัสนิสิต</br>
ให้เขียนโปรแกรมเพื่อจำลองวิธีการที่ได้อธิบายข้างต้น</br>

ข้อมูลนำเข้า</br>
• บรรทัดแรก ระบุจำนวนเต็ม n แทนจำนวนภาควิชา</br>
• n บรรทัดถัดมา ระบุชื่อภาควิชา และจำนวนนิสิตที่แต่ละภาควิชาสามารถรับได้ เป็นจำนวนเต็ม</br>
• บรรทัดถัดมา ระบุจำนวนเต็ม m แทนจำนวนนิสิตที่ทำการเลือกภาควิชา</br>
• m บรรทัดถัดมา ระบุข้อมูลนิสิตแต่ละคน ประกอบด้วยรหัสนิสิต 10 หลัก คะแนนเป็นจำนวนทศนิยม และภาควิชาที่นิสิต</br>
เลือก 4 ภาควิชา เรียงตามความต้องการลำดับที่ 1-4 (รับประกันว่าไม่มีนิสิตที่มีคะแนนเท่ากัน)</br>

ข้อมูลส่งออก</br>
แสดงภาควิชาที่นิสิตแต่ละคนได้ เรียงตามรหัสนิสิต (รับประกันว่านิสิตทุกคนจะมีภาควิชา)</br>

````python
avial={} #{'CP': 1, 'ME': 2, 'PE': 1, 'CHE': 1, 'MT': 3}
for k in range(int(input())): 
    dept,n=input().split()# Dept มีที่ว่างกี่ที่   
    avial[dept]=int(n)# {Dep ที่ว่าง}
s=[]#เดี๋ยวเราจะ sort 
for k in range(int(input())):
    x=input().split() 
    x[0],x[1]=float(x[1]),x[0] #ให้สลับที่เอาคะแนนขึ้นก่อน แปลง float ตามด้วย ID   
    s.append(x) #[[23.6, '59301234521', 'PE', 'CP', 'MT', 'CHE'], ...]

r={}  #result
for e in sorted(s)[::-1] :  #reverse มากไปน้อย  ลุยไปในแต่ละ dep ที่เขาเลือก และเช็คว่ามีที่ว่างไหม
    for d in e[2:]: #เช็ค dep อยู่ index2 เป็นต้นไป
        if avial[d]>0: #{'CP': 1, 'ME': 2, 'PE': 1, 'CHE': 1, 'MT': 3}
            r[e[1]]=d #บันทึกคณะที่ได้ลง result
            avial[d]-=1 #หัก value ออก 1
            break
#แสดงภาควิชาที่นิสิตแต่ละคนได้ เรียงตามรหัสนิสิต (รับประกันว่านิสิตทุกคนจะมีภาควิชา)
for e in sorted(r):
    print(e,r[e])
 
````

10-11: รถไฟฟ้า  </br>
นักท่องเที่ยวชาวต่างชาติคนหนึ่งต้องการเดินทางท่องเที่ยวตามสถานที่สำคัญในกรุงเทพมหานคร เขาจึงเลือกใช้บริการรถไฟฟ้า เขาขึ้น </br>
รถไฟฟ้าที่สถานี A แต่เขามีเงินไม่มาก จึงเดินทางได้ไม่เกิน 2 ช่วงสถานีเท่านั้น เขาอยากทราบว่า จากสถานี A ที่เขาอยู่ตอนนี้ เขาจะ </br>
สามารถไปถึงสถานีใดได้บ้าง </br>
ข้อมูลนำเข้า </br>|
ส่วนแรกจะเป็นการบอกว่า สถานีรถไฟฟ้าคู่ใดอยู่ติดกันบ้าง แต่ละบรรทัดจะมีชื่อสถานี 2 ชื่อ คั่นด้วยเว้นวรรค หมายความว่า สถานี 2 </br>
สถานีนี้อยู่ติดกัน สถานีรถไฟฟ้าทั้งหมดไม่จำเป็นต้องเชื่อมต่อถึงกันก็ได้ (ลองดูในตัวอย่าง) </br>
บรรทัดสุดท้ายมีชื่อสถานี 1 ชื่อ แทนสถานีที่นักท่องเที่ยวชาวต่างชาติอยู่ในตอนนี้ (ซึ่งอาจจะไม่เชื่อมกับสถานีใดเลยก็ได้) </br>
ข้อมูลส่งออก </br>
แสดงชื่อสถานีที่นักท่องเที่ยวชาวต่างชาติสามารถไปได้ (ห่างกันไม่เกิน 2 ช่วงสถานี) ให้ตอบเรียงตามลำดับพจนานุกรม </br>

## ห่างกัน2 ช่วงสถานี มีสถานีไหนบ้าง
```
input 
Siam ChitLom
ChitLom PhloenChit
PhloenChit Nana
Siam NationalStadium
Ratchadamri Siam
Siam PhayaThai
Ratchadamri SalaDaeng
ThongLo Ekkamai
Ekkamai ThongLo
Siam #สถาที่อยู่

output
ChitLom
NationalStadium
PhayaThai
PhloenChit
Ratchadamri
SalaDaeng
Siam
```

````python
#อยู่ถัดไปจากสยาม 2 สถานี
next={}
while True:
    x =input().split()
    if len(x)==1: break
    s1,s2=x #2ตัวแปรมารับ 2สถานนีติดกัน
    #เช็คก่อนว่ายังไม่มี ใน next   สร้าง set ว่าง
    if s1 not in next:next[s1]=set()
    if s2 not in next:next[s2]=set()
    next[s1].add(s2) #เก็บ next  ของ s1 เพิ่ม s2เข้าไป
    next[s2].add(s1) #เก็บ next  ของ s1 เพิ่ม s2เข้าไป

#หาถัดไป 2สถานี 
q=x[0] #ตัวแปร input บรรทัดสุดท้าย siam
if q not in next: #กรณีไม่ติดกับอะไรเลย
    print (q)
else:
    results=set(next[q]) # copy next(q)     ผลลัพธ์  'Siam': {'Ratchadamri', 'PhayaThai'}
    for s in next[q]: #หยิบสถานทีใน 'Siam'  คือ  Ratchadamri', 'PhayaThai'
        results |=next[s] # union set  results กับ  next[s]
    #เรียง และ เว้นบรรทัด
    print('\n'.join(sorted(results)))  
````
