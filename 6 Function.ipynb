06-05: จำนวนเฉพาะถัดไป </br>
ฟังก์ชัน is_prime(n) ข้างล่างนี้ตรวจว่า n ที่ได้รับเป็นจำนวนเฉพาะหรือไม่ จงใช้ is_prime ให้เป็นประโยชน์ในการเขียน </br>
ฟังก์ชัน next_prime(n) และ next_twin_prim(n) ตามคำอธิบายที่เขียนใน comment ของแต่ละฟังก์ชัน </br>

```python
def is_prime(n):
# ทดสอบว่า n เป็นจานวนเฉพาะหรือไม่
    if n <= 1:
        return False
    for k in range(2,int(n**0.5)+1):
        if n%k == 0:
            return False
    return True
 
def next_prime(N): 
    N+=1
    while not is_prime(N): #วนจน true
        N+=1
    return N

# คืนจานวนเฉพาะตัวทมีี่ค่านอ้ ยสุดทมี่ ากกว่า N

def next_twin_prime(N):
    n1=next_prime(N)
    n2=next_prime(n1)
    while n2-n1 !=2:
        n1=n2
        n2=next_prime(n1)
    return n1,n2
        
    
# คืนจานวนเฉพาะสองค่าทเี่ ป็น twin prime ทมีี่ค่านอ้ ยสุดทมี่ ากกว่า N
# twin prime คือจานวนเฉพาะทมีี่ค่าต่างกนั 2 เช่น 11 กับ 13 หรือ 41 กับ 43

exec(input().strip()) # ตอ้ งมีคาสั่งนี้ ตรงนี้ ตอนส่งให้ Grader ตรวจ
```

ข้อมูลนำเข้า
บรรทัดแรกเป็นสตริงที่เก็บลำดับตัวอักษรของเฉลยในแต่ละข้อเรียงกันไป
บรรทัดต่อมาเป็นจำนวนเต็ม N แทนจำนวนนักเรียนที่ส่งคำตอบ
อีก N บรรทัด แต่ละบรรทัดมี 2 สตริงคั่นด้วยช่องว่าง ตัวแรกเป็นเลขประจำตัว อีกตัวป็นสตริงที่เก็บลำดับตัวอักษรของคำตอบ

```python
ข้อมูลส่งออก
รายการของเลขประจำตัว คะแนน และเกรดที่ได้ของนักเรียนแต่ละคน คนละบรรทัด โดยเรียงลำดับตามคะแนนจากมากไปนัอย
ในกรณีที่คะแนนเท่ากัน ให้เรียงตามเลขประจำตัวจากมากไปน้อย (คะแนนที่ได้นั้นเป็นร้อยละของคำตอบที่ถูกต้อง)
```

```python
def read_answers():
    N = int(input()) #รับ input แปลงเป็น int เก็บไว้ใน N
    answers = []
    for k in range(N):#วน N ครั้ง
        sid, ans = input().split() #รับคำตอบ 2222 AAAAB
        answers.append([sid, ans]) # เก็บไว้ใน list [[id,ans] ,[id,ans] ]
    return answers

#ให้คะแนน รับ answer, solution string คืน score int
def marking(answer, solution):
    score = 0
    for i in range(len(answer)):
        if answer[i] == solution[i]:
            score += 1
    return score


#รับ score [0-100] int คืนเกรด string

def grading(score):
    g = [[80,"A"], [70,"B"],
    [60,"C"], [50,"D"]]
    for a,b in g:
        if score >= a:
            return b
    return "F"

def scoring(answers, solution):
    scores = []
    for sid, ans in answers:
        score = marking(ans, solution)/len(solution) * 100
        grade = grading(score)
        scores.append([sid, score, grade])
    return scores

def report(scores):
    for sid,sc,grade in scores:
        print(sid, sc, grade)
    
def sort(scores):
    x = []
    for sid,score,grade in scores:
        x.append([score, sid, grade])
        x.sort()
    for i in range(len(x)):
        scores[i] = [x[i][1], x[i][0],x[i][2]]

# TODO       
sol=input()
answers =read_answers()
scores=scoring(answers,sol)
sort(scores)
report(scores[::-1])

```

06-07: สามฟังก์ชันเกี่ยวกับระยะสั้นสุด </br>
จงเขียนฟังก์ชัน distance1, distance2 และ distance3 ที่ทำงานตามที่เขียนใน comment</br>

```python
def distance1(x1, y1, x2, y2):
    dx=x1-x2
    dy=y1-y2
    d=(dx**2+dy**2)**0.5
    return d    

# คืนระยะห่างระหว่างจุด (x1,y1) กับ (x2,y2)
# ตัวอย่างการใช:้ d1 = distance1(0.0, 0, 3, 4) ได้ d1 = 5.0
def distance2(p1, p2):
    x1=p1[0]
    x2=p2[0]
    y1=p1[1]
    y2=p2[1]
    return distance1(x1, y1, x2, y2)    
# p1 และ p2 เป็นลิสต์
# แต่ละลิสต์แทนจุด ทเี่ ป็นลสิ ต์ทภี่ ายในมี 2 ช่อง เก็บพิกัด x กับ y
# คืนระยะระหว่างจุด p1 กับ p2
# ตัวอย่างการใช:้ d2 = distance2([0.0, 0], [3, 4]) ได้ d2 = 5.0
def distance3(c1, c2):
    #c1[x,y,r]
    d=distance2(c1[:2],c2[:2])
    is_overlap = (d <=c1[2]+c2[2]) #จะตัดกันถ้าระยะห่างน้อยกว่ารัศมีบวกกัน
    return d,is_overlap

    
# c1 และ c2 แทนวงกลม 2 วง
# แต่ละลิสต์เป็นลิสต์ 3 ช่อง เก็บพิกัด x กับ y (ของจุดศูนย์กลาง) และรัศมี
# คืนระยะระหว่างจุดศูนย์กลางของ c1 กับ c2 และค่าจริง/เท็จว่า c1 กับ c2 แตะหรือทับกันหรือไม่
# ตัวอย่างการใช:้ d3,overlap = distance3([0.0, 0, 1], [5, 0, 2])
# ได้ d3 = 5.0, overlap = False
def perimeter(points):
    #[[x,y],[x,y]...]
    p=0
    for i in range(len(points)):
        p+=distance2(points[i],points[i-1])
    return p
    
# points เป็นลิสต์ของจุดต่าง ๆ แต่ละจุดเป็นลสิ ต์ 2 ช่อง (เก็บพิกัด x และ y)
# จุดเหล่านี้คือลาดับของมุมของรูปหลายเหลยี่ ม (รูป k เหลยี่ มก็มี k จุด, k>=3)
# คืนความยาวรอบรูปของรูปหลายเหลยี่ มทกี่ าหนดโดย points
exec(input().strip()) # ตอ้ งมีคาสั่งนี้ ตรงนี้ ตอนส่งให้ Grader ตรวจ
```
