```python
#Start time
# input 1 ชั่วโมง 0-23
# input2 นาที 0-59
#input 3 วินาที 0-59
#End time 
#input 4 ชั่วโมง 0-23
#input5 นาที 0-59
#input6 วินาที 0-59
#out put ช่วงเวลา
h1=int(input("ชั่วโมง"))
m1=int(input("นาที"))
s1=int(input("วินาที"))
h2=int(input("ชั่วโมง"))
m2=int(input("นาที"))
s2=int(input("วินาที"))
t1 = h1*60*60 + m1*60 + s1 #แปลงเป็นวินาที
t2 = h2*60*60 + m2*60 + s2
dt = t2 - t1
dh = dt // (60*60)#แปลงผลต่างวินาทีเป็นชั่วโมง แล้วปัดเศษทิ้ง
dt -= dh * 60*60#หักชั่วโมงออก
dm = dt // 60#แปลงวินาทีที่เหลือเป็นนาที
dt -= dm*60#หักนาทีออก
ds = dt#วินาทีที่เหลือ
print(str(dh) + ":" + str(dm) + ":" + str(ds))
```

แก้ไขสูตร
1วันมี 24 ชม เอาไป %mod
- เริ่ม 2 จบ 4 dh = ((4-2)+24) = 26 ผิด จัดการด้วย %24
- เริ่ม 2 จบ 4 dh = ((4-2)+24)%24 = 2

นาที
ทำคล้ายๆกัน 1 รอบวันมี sec_in_day 24x60x60 วินาที เอาไป % mod
- dt = (t2-t1+sec_in_day)%sec_in_day

```python
#แก้ไขสูตร
h1=int(input("ชั่วโมง"))
m1=int(input("นาที"))
s1=int(input("วินาที"))
h2=int(input("ชั่วโมง"))
m2=int(input("นาที"))
s2=int(input("วินาที"))
t1 = h1*60*60 + m1*60 + s1 #แปลงเป็นวินาที
t2 = h2*60*60 + m2*60 + s2
sec_in_day= 24*60*60
dt = (t2 - t1+sec_in_day)%sec_in_day
# -1%24 =3 ,  dt = (t2 - t1)%sec_in_day ใช้แบบนี้ได้เหมือนกัน
dh = dt //(60*60)#แปลงผลต่างวินาทีเป็นชั่วโมง แล้วปัดเศษทิ้ง
#dt -= dh * 60*60#หักชั่วโมงออก
dt = dt%3600 #ตัดที่เกิน ชั่วโมงออก
dm = dt // 60#แปลงวินาทีที่เหลือเป็นนาที
#dt -= dm*60#หักนาทีออก
ds = dt%60#วินาทีที่เหลือ
print(str(dh) + ":" + str(dm) + ":" + str(ds))
```
