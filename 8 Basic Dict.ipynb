08-04: ยอดขายไอศกรีม </br>
จงเขียนโปรแกรมอ่านราคาไอศกรีมต่าง ๆ ตามด้วยการขายสินค้าต่าง ๆ ในร้าน</br>
เพื่อหาและแสดงยอดขายรวมของไอศกรีม และไอศกรีมชนิดที่มียอดขายสูงสุด</br>
ข้อมูลนำเข้า</br>
บรรทัดแรกเป็นจำนวนเต็ม N</br>
N บรรทัดต่อมาเป็นข้อมูลราคาขายของไอศกรีม ประกอบด้วย ชื่อไอศกรีม ตามด้วยราคา (จำนวนจริง) คั่นด้วยช่องว่าง</br>
บรรทัดที่ N+2 เป็นจำนวนเต็ม M</br>
M บรรทัดต่อมาเป็นการขายสินค้า ประกอบด้วย ชื่อสินค้า ตามด้วยจำนวนที่ขายได้ คั่นด้วยช่องว่าง</br>
ข้อมูลส่งออก</br>
ถ้าขายไอศกรีมไม่ได้เลย ให้แสดงข้อความว่า No ice cream sales</br>
ถ้ามีการขายไอศกรีม ให้หาและแสดงยอดขายรวมของไอศกรีมในบรรทัดหนึ่ง และตามด้วยรายการของชื่อไอศกรีมที่มียอดขายสูงสุด</br>
อีกบรรทัดหนึ่ง ถ้ามีหลายชื่อให้แสดงชื่อเรียงตามพจนานุกรม (ดูรูปแบบการแสดงในตัวอย่าง)</br>

````python
price ={}
for k in range(int(input())): #กำหนดบรรทัดที่จะตามมา
    ic,p=input().split() #รับ key value
    price[ic]=float(p) #เก็บ price
sales={}
total=0
for k in range(int(input())):  #กำหนดบรรทัดที่จะตามมา
    prod,k=input().split() # รับข้อมูล prod key  price value
    if prod in price: #ถามว่า prod ใช่ ice รึเปล่า อยู่ใน price
        p = price[prod]*int(k) #ราคา x ยอดขาย = sales
        if prod in sales: #ถามว่า prod อยู่ใน sales หรือยยัง ถ้าอยู่จริง
            sales[prod]+=p #บวกเพิ่ม
        else:
            sales[prod]=p # ถ้าไม่อยู่ใส่ p เข้่าไป
        total+=p #ยอดรวม
if total==0: #ถ้าไม่มียอดขาย
    print("No ice cream salse")
else:
    x=[] #เก็บเป็น list [[s,p],...]
    for ic in sales:
        x.append([sales[ic],ic])
    maxsales=max(x)[0]  #หา max ของ x 
    mx=[]
    for s,ic in x: #แต่ละ sales ใน x
        if s==maxsales:
                  mx.append(ic)
            
    mx.sort() #เรียงลำดับ
    print('Total ice cream sales:',total)
    print('Top sales:',','.join(mx))
        
````

````python
#text to key
T2K={"a":"2","b":"22","c":"222",
     "d":"3","e":"33","f":"333",
    "g":"4","h":"44","i":"444",
     "j":"5","k":"55","l":"555",
     "m":"6","n":"66","o":"666" ,
   "p":"7", "q":"77", "r":"777",
"s":"7777", "t":"8", "u":"88", "v":"888", "w":"9",
    "x":"99", "y":"999", "z":"9999" , " ":'0'}
K2T={}
for k in T2K:
    v=T2K[k]
    K2T[v]=k
def text2keys( text ):
    keys=' '
    for c in text.lower():
        if c in T2K:
            keys+=T2K[c]+ ' '
    return keys.strip()
def keys2text( keys ):
    t =' '
    for k in keys.split():
        t+=K2T[k] 
    return t
exec(input().strip())
````

กำหนดให้เราแทนเงินด้วย dict ที่มี key เป็นมูลค่าของเหรียญหรือธนบัตร และ value คือจำนวนเหรียญหรือธนบัตรที่มี เช่น </br>
{100: 5, 50: 2, 10: 5, 1: 15} แทนธนบัตรหนึ่งร้อยบาท 5 ใบ, ห้าสิบ 2 ใบ, สิบบาท 5 ใบ และเหรียญบาท 15 </br>
เหรียญ </br>
จงเขียนฟังก์ชันต่าง ๆ ข้างล่างนี้ (ดูตัวอย่างข้างล่างประกอบ) </br>
- total(pocket) คืนจำนวนเงินรวมใน pocket</br>
- take(pocket, money) เติมเงินจาก money เข้าใน pocket (ทั้งคู่เป็น dict ที่เก็บเงิน)</br>
- pay(pocket, amt) ตัดเงินใน pocket เป็นจำนวน amt (เป็นจำนวนเต็ม) ตัวฟังก์ชันจะคืนเงิน (เป็น dict) ที่จ่าย</br>
ออกไป ในกรณีที่จ่ายเป็นจำนวน amt ไม่ได้ ให้คืน dict ว่าง</br>
โดยการจ่ายเงิน ใช้วิธีการเลือกธนบัตรหรือเหรียญที่มีในกระเป๋าที่มีมูลค่าสูงสุดที่จ่ายได้ก่อน เช่น {100: 5, 50: 2, 10:5,</br>
1: 15} ต้องการจ่ายออก 57 ก็จะหยิบ 50 ออกหนึ่งใบ และ 1 ออกเจ็ดเหรียญ</br>
```

def pay(pocket, amt):

```
ข้อมูลนำเข้า</br>
คำสั่งภาษา Python ที่ใช้ทดสอบการทำงานของฟังก์ชัน</br>
ข้อมูลส่งออก</br>
ผลที่ได้จากการสั่งทำงานคำสั่งที่ได้รับ</br>

````python
def total(pocket):  #คืนจำนวนเงินรวมใน pocket
    s=0
    for m in pocket:
        s+=m*pocket[m]
    return s
def take(pocket, money_in):  #เติมเงินจาก money เข้าใน pocket
     for m in money_in:
            if m in pocket: #pocket มี m ไหมถ้ามีเพิ่มเลย ถ้าไม่มีเติมเข้าไป
                pocket[m]+=money_in[m]    
            else: 
                pocket[m] =money_in[m] 
        
def pay(pocket, amt): #ตัดเงินใน pocket เป็นจำนวน amt (เป็นจำนวนเต็ม) ตัวฟังก์ชันจะคืนเงิน (เป็น dict) ที่จ่าย 
#ออกไป ในกรณีที่จ่ายเป็นจำนวน amt ไม่ได้ ให้คืน dict ว่าง
    money_out={}
    #มูลค่าของเงินเรียงจากน้อยไปมาก
    for m in sorted(pocket.keys())[::-1] :#revers เรียงจากมากไปน้อย
        if amt>=m: #เรามีพอทอนไหม #amt =330 , m=100 , pocket[m]=4 , amt//m ใช้กี่ใบ
            c=min(pocket[m], amt//m)#ค่าไหนน้อยกว่า ก็จ่ายค่านั้น
            money_out[m]=c #แบงค์ที่จ่าย
            pocket[m]-=c#หักแบงค์ออก
            amt-=m*c#หักมูลค่าเหลือออก
        if amt>0: #จ่ายไม่ได้เอาคืนกลับเข้าไปใน pocket
            take(pocket,money_out)
            money_out={}
      
        return money_out
 
exec(input().strip()) 
````
